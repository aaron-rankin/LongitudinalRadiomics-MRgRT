```python

import pandas as pd
# create a bar chart of number of features per ICC group
df_ICC = pd.read_csv('../Output/Submission/features/Manual_ICC_Values.csv')

# group by feature and calculate mean ICC
groups = df_ICC['ICC_Class'].unique()
# ICC < 0.5 = Poor, 0.5 <= ICC < 0.75 = Moderate, 0.75 <= ICC < 0.9 = Good, ICC >= 0.9 = Excellent


df_ICC_grouped = df_ICC.groupby('Feature')['ICC'].mean().reset_index()
df_ICC_grouped['ICC_Class'] = pd.cut(df_ICC_grouped['ICC'], bins=[0, 0.5, 0.75, 0.9, 1], labels=['Poor', 'Moderate', 'Good', 'Excellent'])

# count the number of features in each group
feature_count = df_ICC_grouped['ICC_Class'].value_counts().reset_index()
feature_count.columns = ['ICC_Class', 'Count']

x_vals = feature_count['Count']
y_vals = feature_count['ICC_Class']

# create a bar chart
plt.figure(figsize=(6, 6), facecolor='white')

sns.set_style('white')
# border aroun
sns.barplot(y=x_vals, x=y_vals, palette='PuOr', orient='v')
plt.xlabel('')

# turn all the spines off
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.yticks([])
plt.ylabel('')

plt.xticks(labels=['Poor', 'Moderate', 'Good', 'Excellent'], rotation=0, ticks=[0, 1, 2, 3],
           fontsize=12)

plt.savefig('C:\\Users\\b01297ar\\Documents\\ProstateMRL-local\\Paper1-Write\\Figures\\Diagram/ICCChart-v2.png', dpi=300, bbox_inches='tight')
```

```python
# correlate volume to feature value
df_fts = pd.read_csv('../Output/Submission/features/features_all.csv')
df_fts = df_fts[df_fts['Fraction'] == 4]
df_vol = df_fts[df_fts['Feature'] == 'shape_MeshVolume']

df_ft = df_fts[df_fts['Feature'] == 'glrlm_RunLengthNonUniformity']


vals_ft = df_ft['FeatureValue'].values
vals_vol = df_vol['FeatureValue'].values

# create a scatter plot
plt.figure(figsize=(6, 6))
sns.set_style('darkgrid')
sns.scatterplot(x=vals_vol, y=vals_ft, color='tab:orange', edgecolor='white', linewidth=0.5, alpha=0.8, marker='o'
                , s=100)
plt.xlabel('Volume', fontsize=16, labelpad=10)
plt.ylabel('Feature', fontsize=16, labelpad=10)

# turn all the spines off
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# turn off the ticks and labels
plt.tick_params(axis='both', which='both', bottom=False, top=False, labelbottom=False, right=False, left=False, labelleft=False)

# make axis thicker and grey
plt.gca().spines['left'].set_linewidth(3)
plt.gca().spines['left'].set_color('grey')
plt.gca().spines['bottom'].set_linewidth(3)
plt.gca().spines['bottom'].set_color('grey')

# pad the axis labels



plt.savefig('C:\\Users\\b01297ar\\Documents\\ProstateMRL-local\\Paper1-Write\\Figures\\Diagram\\VolumeScatter.png', dpi=300, bbox_inches='tight')

```
