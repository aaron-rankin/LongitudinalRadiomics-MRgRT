```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
csvs = os.listdir('..\\TestData\DMs2\\')
```

```python

for pat in csvs:
    df_dist = pd.read_csv('..\\TestData\DMs\csvs\\' + pat, index_col=0)
    features = df_dist.columns
    
    fig, ax = plt.subplots(figsize=(10,10))
    sns.heatmap(df_dist, cmap="viridis", ax=ax)
    plt.title('Euclidean Distance Map', fontsize=16)
    # remove ticks on colorbar
    cbar = ax.collections[0].colorbar
    cbar.set_ticks([])

    # make sure all ticks show
    # plt.xticks(np.arange(len(features)) + 0.5, features, fontsize=6)
    # plt.yticks(np.arange(len(features)) + 0.5, features, fontsize=6)
    
    # set no ticks
    plt.xticks([])
    plt.yticks([])


    plt.show()
```
