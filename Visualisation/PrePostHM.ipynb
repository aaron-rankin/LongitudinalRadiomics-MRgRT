```python
import SimpleITK as sitk
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


key = pd.read_csv('C:/Users/b01297ar/Documents/ProstateMRL-local/ProstateMRL-Radiomics/Paper1-Release/Input/Default/Default_EXtractionKey_T.csv')
patIDs = key['PatID'].unique()

for pat in patIDs:
    key_pat = key[key['PatID'] == pat]

    img_paths = key_pat['ImagePath'].values

    for img_path in img_paths:
        img_path = img_path.replace('HM-FSTP', 'HM-FS')
        img = sitk.ReadImage(img_path)
        img = sitk.GetArrayFromImage(img)
        
        # plot histogram
        plt.hist(img.flatten(), bins=50)
        plt.title(img_path)
        plt.show()
```
